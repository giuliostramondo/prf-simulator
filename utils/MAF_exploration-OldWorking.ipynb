mapping scheme space exploration

1) define a grammar:
	there are two kind of symbles : operations * / + - floor(A) %
									variables i j p q
	those define a valid equation EQ

2) define module assignment function:
	l =  (EQ1)%p
	k = (EQ2)%q

3) define conflicts goal:
	no_conflicts=(i+1,j+1),(i+1,j+2),(i+1,j+3),(i+1,j+1),(i+2,j+1),(i+3,j+1) // specify a cross shape

4) set p and q
5) set M and N

6) generate two equations given the grammar called EQ1 and EQ2 (http://www.nltk.org/howto/generate.html)

7) Use EQ1 and EQ2 to generate a MAF table for 0<i<M 0<j<N (python eval)

8) check compliance with conflicts goal
	for 0<i<M-MaxOffsetonI 0<j<N-MaxOffsetOnJ:
		for no_conflict1 in no_conflicts:
			for no_conflict2 in no_conflicts:
				if(no_conflict1!=no_conflict2):
					if (MAF_TABLE[no_conflict1] == MAF_TABLE[no_conflict2]):
						goto step 6
		#found right MAF
		return l and k

```python
def gen_collisions(i,j):
    no_collisions_cross=[(i,j+1),(i+1,j),(i+1,j+1),(i+1,j+2),(i+2,j+1)]
    return no_collisions_cross

def check_collisions(MAF_Table,gen_collisions,debug=False):
    for i in range(N):
        for j in range(M):
            no_collisions = gen_collisions(i,j)
            #Check collisions within matrix NxM
            skip = 0
            for item in no_collisions:
                if item[0] >= N or item[1]>=M:
                    skip=1
                    break
            if skip:
                continue
            for item1 in no_collisions:
                for item2 in no_collisions:
                    if item1 != item2:
                        try:
                            if MAF_Table[item1[0]][item1[1]] == MAF_Table[item2[0]][item2[1]]:
                                #fordebug
                                if debug:
                                    print str(item1[0])+","+str(item1[1])+"=="+str(item2[0])+","+str(item2[1])
                                    print "i="+str(i)+" j="+str(j)
                                    print no_collisions
                                return 0
                        except:
                            print MAF_Table
                            print str(len(MAF_Table))
                            print str(item1[0])+" "+str(item1[1])
                            print str(item2[0])+" "+str(item2[1])
                            raise
    return 1

def generate_MAF_Table(N,M,mv,mh):
    MAF_Table = [[(0,0) for x in range(M)] for y in range(N)]
    for i in range(N):
        for j in range(M):
            try:
                MAF_Table[i][j]=(eval(mv),eval(mh))
            except:
                return []
    return MAF_Table

def find_MAF(eq_list,gen_collisions,N,M,p,q,progress=False):
    gen_collisions
    if progress:
        i=0
        tot=len(eq_list)
        for EQ1 in eq_list:
            print str(i/float(tot))+"%"
            for EQ2 in eq_list:
                mv=EQ1+"%p"
                mh=EQ2+"%q"
                MAF_Table= generate_MAF_Table(N,M,mv,mh)
                if not MAF_Table:
                    continue
                if check_collisions(MAF_Table,gen_collisions):
                    print "mv="+mv
                    print "mh="+mh
                    return 1
            i=i+1
        return 0
    else:
        for EQ1 in eq_list:
            for EQ2 in eq_list:
                mv=EQ1+"%p"
                mh=EQ2+"%q"
                MAF_Table= generate_MAF_Table(N,M,mv,mh)
                if not MAF_Table:
                    continue
                if check_collisions(MAF_Table,gen_collisions):
                    print "mv="+mv
                    print "mh="+mh
                    return 1
        return 0
```
